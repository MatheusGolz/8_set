```python
class Base:
    ''' Base do pacote, reponsável pela entrada de dados e informações gerais'''
    def __init__(self,filename):
        import pandas as pd
        self.filename = filename + '.csv'
        self.df = pd.read_csv(self.filename)

    def fct_resumo(self):
        print('\nO arquivo é composto dos seguintes dados:')
        print(self.df.columns)
        print('\nA quantidade de amostras encontradas foi de:')
        print(self.df.count())


    def fct_info(self):
        data_info = self.df.info()
        print(data_info)

    def fct_null(self):
        null_Van = self.df['Van'].isna().sum()
        null_Vbn = self.df['Vbn'].isna().sum()
        null_Vcn = self.df['Vcn'].isna().sum()
        print('\n Número de elementos nulos na fase A: ' + str(null_Van))
        print(' Número de elementos nulos na fase B: ' + str(null_Vbn))
        print(' Número de elementos nulos na fase C: ' + str(null_Vcn))

    def fct_biggest(self):
        print('\nMaior valor da fase A: ' + str(self.df['Van'].max()))
        pos = self.df.loc[self.df['Van'].idxmax()]
        print('A maior amostra foi registrada: ' + pos[0] + '. Hora: ' + pos[1])
        print('Menor valor da fase A: ' + str(self.df['Van'].min()))
        pos = self.df.loc[self.df['Van'].idxmin()]
        print('A menor amostra foi registrada: ' + pos[0] + '. Hora: ' + pos[1])

        print('\nMaior valor da fase B: ' + str(self.df['Vbn'].max()))
        pos = self.df.loc[self.df['Vbn'].idxmax()]
        print('A maior amostra foi registrada: ' + pos[0] + '. Hora: ' + pos[1])
        print('Menor valor da fase B: ' + str(self.df['Vbn'].min()))
        pos = self.df.loc[self.df['Vbn'].idxmin()]
        print('A menor amostra foi registrada: ' + pos[0] + '. Hora: ' + pos[1])

        print('\nMaior valor da fase C: ' + str(self.df['Vcn'].max()))
        pos = self.df.loc[self.df['Vcn'].idxmax()]
        print('A maior amostra foi registrada: ' + pos[0] + '. Hora: ' + pos[1])
        print('Menor valor da fase C: ' + str(self.df['Vcn'].min()))
        pos = self.df.loc[self.df['Vcn'].idxmin()]
        print('A menor amostra foi registrada: ' + pos[0] + '. Hora: ' + pos[1])
```

```python
inst = Base('RMSdata')
```

```python
inst.fct_info()
```

```python
inst.fct_resumo()
```

```python
inst.fct_null()
```

```python
inst.fct_biggest()
```
